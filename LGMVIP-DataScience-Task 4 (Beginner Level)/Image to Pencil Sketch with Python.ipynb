For converting an image into a pencil sketch with Python, I am going to use OpenCV library. I will complete this task in a step by step process and finally I will define a function where you can give path of an image and you will get its pencil sketch in return.

Source of images: https://www.wallpapertip.com/

## Import necessary libraries

```python
import cv2 #OpenCV
import matplotlib.pyplot as plt #to display image 
```

## Step 1: Find an image that you want to convert to a pencil sketch
I am going to use this dragon image for making pencil sketch.

```python
path = "dragon.jpg"
img = cv2.imread(path)
plt.imshow(img)
```

## Step 2: Read the image in RGB format

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

## Step 3: Convert image to a grayscale image
This will turn an image into a classic black and white photo.

```python
gray_img = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
plt.imshow(gray_img, cmap='gray')
```

## Step 4: Invert the grayscale image 
This will turn an image into inverted grayscale image, also called negative image. 
Inversion is done to enhance details. It converts white spaces into black and black spaces into white.

```python
inverted_img = 255 - gray_img
plt.imshow(inverted_img, cmap = 'gray')
```

## Step 5: Blur the image by using the Gaussian Function in OpenCV

```python
blurred = cv2.GaussianBlur(inverted_img, (21, 21), 0)
```

## Step 6: Invert the blurred image

This is the final step where we get the pencil sketch of the image.

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_img, inverted_blurred, scale=256)
plt.imshow(pencil_sketch, cmap = 'gray')
```

## View original image and its pencil sketch together for better comparison

```python
plt.figure(figsize=(11,4))
plt.subplot(1,2,1)
plt.imshow(RGB_img)
plt.subplot(1,2,2)
plt.imshow(pencil_sketch, cmap='gray')
```

## Define a function which can give us pencil sketches of given images

```python
def pencil_sketch_creator(path):
    img = cv2.imread(path) #select image
    RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #read in RGB format
    gray_img = cv2.imread(path,cv2.IMREAD_GRAYSCALE) #convert it into grayscale
    inverted_img = 255 - gray_img #invert grayscale image
    blurred = cv2.GaussianBlur(inverted_img, (21, 21), 0) #blur inverted image
    inverted_blurred = 255 - blurred #invert blurred image
    pencil_sketch = cv2.divide(gray_img, inverted_blurred, scale=256) #pencil sketch of image
    plt.subplot(1,2,1)
    plt.imshow(RGB_img) #original image
    plt.subplot(1,2,2)
    plt.imshow(pencil_sketch, cmap='gray') #pencil sketch
```
